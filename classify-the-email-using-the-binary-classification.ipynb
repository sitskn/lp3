ML B-2 [Classify the email using the binary classification method.]

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier  
from sklearn import metrics
```

```python
df = pd.read_csv('/Users/shreyas/Downloads/RAW/codes/emails.csv')
df
```

```python
df.shape
```

```python
df.isnull().any()
```

```python
df.drop(columns='Email No.', inplace=True)
df
```

```python
df.columns
```

```python
df.Prediction.unique()
```

```python
df['Prediction'] = df['Prediction'].replace({0:'Not spam', 1:'Spam'})
```

```python
df
```

# KNN

```python
X = df.drop(columns='Prediction',axis = 1)
Y = df['Prediction']
```

```python
X.columns
```

```python
Y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
KN = KNeighborsClassifier
knn = KN(n_neighbors=7)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
print("Prediction: \n")
print(y_pred)
```

```python
# Accuracy

M = metrics.accuracy_score(y_test,y_pred)
print("KNN accuracy: ", M)
```

```python
C = metrics.confusion_matrix(y_test,y_pred)
print("Confusion matrix: ", C)
```

# SVM Classifier

```python
model = SVC(C = 1)   # cost C = 1

model.fit(x_train, y_train)

y_pred = model.predict(x_test)      # predict
```

```python
kc = metrics.confusion_matrix(y_test, y_pred)
print("SVM accuracy: ", kc)
```
